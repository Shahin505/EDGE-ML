```python
from sklearn.model_selection import train_test_split #
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.30,random_state=42) #divide testing and training data.
```

```python
from sklearn.ensemble import RandomForestClassifier #import randomforest algorithm.
rclf=RandomForestClassifier() #create object.
rclf.fit(xtrain,ytrain) #fit the data
rclf.score(xtest,ytest) #check  score.
```

```python
from sklearn.tree import DecisionTreeClassifier #import DecisionTree algorithm.
rlf=DecisionTreeClassifier() #create object
rlf.fit(xtrain,ytrain) #fit the data.
rlf.score(xtest,ytest) #check score.
```

```python
from sklearn.ensemble import GradientBoostingClassifier #import GradientBoosting.
clf = GradientBoostingClassifier(n_estimators=1000, learning_rate=1.0,    max_depth=5, random_state=0)
clf.fit(xtrain,ytrain)
clf.score(xtest,ytest)
```

```python
from xgboost import XGBClassifier  # import XGB
classifier = XGBClassifier() #create object
classifier.fit(xtrain,ytrain) #fit data for XGB
classifier.score(xtest,ytest)
```

```python
from sklearn import svm
cllf = svm.SVC()
cllf.fit(xtrain,ytrain)
cllf.score(xtest,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(xtrain,ytrain)
neigh.score(xtest,ytest)
```
